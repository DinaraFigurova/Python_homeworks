Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
# Гауссовское распределение

x = np.linspace(0,6,100)

mu = list(range(1, 6, 1))
sigma = np.array(range(2, 15, 3))/10

mu_sigma = dict(zip(mu, sigma))
mu_sigma

norm_rv = []

for mu, sigma in mu_sigma.items():
    norm_rv.append(sts.norm(loc=mu, scale=sigma).pdf(x))

    
for norm_rvi in norm_rv:
    plt.plot(x, norm_rvi)
    
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title(f'Плотность Гауссовского распределения при параметрах mu, sigma = {mu_sigma}')

plt.show()
```

```python
# Экспоненциальное распределение

x = np.linspace(0,5,1000)

lam = np.array(range(2, 20, 4))/10

expon_rv = []

for lam_i in lam:
    expon_rv.append(sts.expon(lam_i).pdf(x))

    
for expon_rvi in expon_rv:
    plt.plot(x, expon_rvi)
    
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title(f'Плотность Экспоненциального распределения при лямбда = {lam}')

plt.show()
```

```python
# распределение Cтьюдента

x = np.linspace(0,5,1000)

n = np.array(range(-80, 70, 30))

t_rv = []

for n_i in n:
    t_rv.append(sts.t(n_i).pdf(x))

    
for t_rvi in t_rv:
    plt.plot(x, t_rvi)
    
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title(f'Плотность распределения Стьюдента при n = {n}')

plt.show()
```
