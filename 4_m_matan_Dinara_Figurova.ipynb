*Уровень 0:*

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.misc import derivative
from sympy import diff, symbols, cos, sin
```

```python
# Определяем функцию f(x) = x^5 + 4sin(2x) + cos(3x+3):
    
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)
```

```python
# Визуализируем функцию:
    
x = np.array([i for i in range(-200, 200)])
y = np.array([f(x_i) for x_i in x])

plt.title('Функция f(x) = x^5 + 4sin(2x) + cos(3x+3)',fontsize=10)

plt.plot(x, y)
```

```python
# Первая производная в точке x = 1:

derivative(f, 1.0, dx=1e-6)
```

```python
# Вторая производная в точке x = 1:

derivative(f, 1.0, dx=1e-6, n = 2)
```

```python
# Первая производная:

x, y = symbols('x y')
diff((sin(2*x+1))**5, x)
```

```python
# Визуализируем первую производную на отрезке [-5; 5]:
    
x = np.array([i for i in range(-5, 5)])
y = np.array([(10*sin(2*x_i+1)*cos(2*x_i+1)) for x_i in x])

plt.title('Функция 10sin4(2x+1)cos(2x+1) на отрезке [-5; 5]',fontsize=10)

plt.plot(x, y)
```

```python
# Вторая производная:

x, y = symbols('x y')
diff(10*sin(2*x+1)*cos(2*x+1), x)
```

```python
# Визуализируем вторую производную на отрезке [-5; 5]:
    
x = np.array([i for i in range(-5, 5)])
y = np.array([(-20*sin(2*x_i+1)**2+20*cos(2*x_i+1)**2) for x_i in x])

plt.title('Функция −20sin2(2x+1)+20cos2(2x+1) на отрезке [-5; 5]',fontsize=10)

plt.plot(x, y)
```
